```
car_rentals_fact_df = session.table("rentals_fact")

car_rentals_fact_df.show()
```

```
from snowflake.snowpark.functions import sum, avg, col

agg_df = car_rentals_fact_df.group_by("CUSTOMER_KEY").agg(
    sum(col("AMOUNT")).alias("total_amount_paid_so_far"),
    avg(col("AMOUNT")).alias("average_amount_paid_so_dar")
)

agg_df.show()
```

```
agg_df.write.save_as_table("rental_cust_agg", mode="overwrite")
```
